# Employee Demographics & Compensation Analysis - EDA
This notebook covers the Exploratory Data Analysis of the generated employee dataset, exploring demographic features and uncovering insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
df = pd.read_csv('../data/cleaned_data.csv')
df.head()
```

```python
# Summary Statistics
df.describe()
```

## 1. Salary Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Salary'], bins=30, kde=True)
plt.title('Salary Distribution')
plt.show()
```

## 2. Department-wise Salary Comparison

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary by Department')
plt.show()
```

## 3. Top 3 Underperforming Departments

```python
dept_perf = df.groupby('Department')['Performance_Rating'].mean().sort_values()
print("Top 3 underperforming departments by performance rating:")
print(dept_perf.head(3))
```
